```python
import pandas as pd
import requests 
from pandas import json_normalize
```

#### Nager.date (https://date.nager.at/swagger/index.html)

We use json_normalize and we can obtain a dataframe with a list of the public holidays worldwide for the next 7 days.

```python
response = requests.get('https://date.nager.at/api/v3/NextPublicHolidaysWorldwide')
print("dates", response.status_code)
```

```python
# response.content
```

```python
import json
dates = response.json()
```

```python
# dates
```

```python
flattened_dates = json_normalize(dates)
```

```python
flattened_dates
```

#### eBird (https://documenter.getpostman.com/view/664302/S1ENwy59#auth-info-124a9625-7cc9-48d5-be64-bb4fa0d59f46).

For a list of birth observations.

First, we needed to create an account and obtain an API token.

```python
api_token = input('Enter ebird api token: ')
```

```python
url = 'https://api.ebird.org/v2/data/obs/KZ/recent'
headers = {
    'X-eBirdApiToken': api_token}

# Realiza la solicitud GET con el encabezado personalizado
response = requests.get(url, headers=headers)
```

```python
# response.content
```

```python
import json
observs = response.json()
```

```python
# observs
```

```python
observs_flatt = json_normalize(observs)
```

We use json_normalize and obtain a list of observations. We have the name and number of species seen, the location with latitude and longitude, etc.

```python
observs_flatt
```

#### Art Institute of Chicago (https://api.artic.edu/docs/#introduction)

For a list of art pieces in the Art Institute of Chicago.

```python
response = requests.get('https://api.artic.edu/api/v1/artworks')
print("chicago", response.status_code)
```

```python
# response.content
```

```python
import json
pieces = response.json()
```

We get 12 pieces of art per page. We have the link for the next page, so we could be iterating for retreiving all the info.

```python
# pieces
```

In this case, we need to use json_normalize on the key 'data'. That's where we find the info of each piece of art.

```python
pieces_flatt = json_normalize(pieces['data'])
```

We get a lot of information for each piece. Title, artist, dates, size, descriptions, etc.

```python
pieces_flatt['artist_title']
```
